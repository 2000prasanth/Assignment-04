```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('C:/Users/USER/Desktop/DSA course/Sales_add.csv')
```

```python
#Check increase in sales after digital marketing initative is taken
```

```python
sales_before = df['Sales_before_digital_add(in $)']
sales_after = df['Sales_After_digital_add(in $)']
months = df['Month']

plt.barh(months, sales_before, label='Sales Before')
#plt.barh(months, sales_after, label='Sales After')

plt.title('Sales Before Digital Advertising')
plt.xlabel('Sales')
plt.ylabel('Month')

plt.legend()

plt.show()
```

```python
plt.barh(months, sales_after, label='Sales After')

plt.title('Sales After Digital Advertising')
plt.xlabel('Sales')
plt.ylabel('Month')

plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots()


before_digital = df['Sales_before_digital_add(in $)']
mean_before = before_digital.mean()
std_before = before_digital.std()
before_digital.plot.kde(ax=ax, label='Sales before digital adv')


after_digital = df['Sales_After_digital_add(in $)']
mean_after = after_digital.mean()
std_after = after_digital.std()
after_digital.plot.kde(ax=ax, label='Sales after digital adv')


ax.legend()


plt.title('Sales before and after digital marketing')
plt.xlabel('Sales')
plt.ylabel('Density')


plt.show()
```

```python
# There as been a considerable increase in sales after digital marketing initiative by the company, but for some months the gap is not considerable
```
